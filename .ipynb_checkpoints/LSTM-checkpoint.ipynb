## 3.1 对天天基金网站上基金类型及数量的分析

### 3.1.1 对各基金类型数量及所占比例进行分析
本小组利用了天天基金网所提供的基金信息数据，网址为http://fund.eastmoney.com/js/fundcode_search.js
利用requests、re两个包提取获得所有基金的基本数据，利用pandas将其转换为DataFrame结构，之后导出到Excel工作簿保存。

```python
import numpy as np
import matplotlib.pyplot as plt
import re
import requests
import pandas as pd

url = 'http://fund.eastmoney.com/js/fundcode_search.js'
num = requests.get(url)

# 通过正则表达式获取基金信息
text = re.findall(r'"(\d*?)","(.*?)","(.*?)","(.*?)","(.*?)"', num.text)

# 转化为DataFrame
基金代码 = []
基金简拼 = []
基金名称 = []
基金类型 = []
基金全拼 = []

for i in text:
    content = list(i)
    基金代码.append(content[0])
    基金简拼.append(content[1])
    基金名称.append(content[2])
    基金类型.append(content[3])
    基金全拼.append(content[4])

基金信息 = pd.DataFrame({'代码': 基金代码, '简拼': 基金简拼, '名称': 基金名称, '类型': 基金类型, '全拼': 基金全拼})

# 写入到Excel文件中
writer = pd.ExcelWriter(r'.\基金数据导出.xlsx', engine='openpyxl')
基金信息.to_excel(writer, sheet_name='基金信息', index=False)
writer.close()  # 关闭writer对象即可
```

之后利用pandas的groupby函数对整体基金的数据进行分组整理，按照类型分组，按计数聚合，最后按照基金数量降序排列，用类型作为索引。通过自定义计算添加列“比例”，计算各类基金在市场上所占比例，得到表一。该部分代码如下：

```python
# 对现有的基金的基本情况进行一个分组统计
group_number=基金信息.groupby('类型').agg(基金数量=('类型','count')).sort_values(by='基金数量',ascending=False).reset_index('类型')

# 计算每种基金所占的比例
group_number['比例']= group_number['基金数量']/group_number['基金数量'].sum()
group_number
```

同时，为了更加直观的展示现有基金市场的各类型基金数量，本小组对基金类型数量进行了统计图表展示和词云图展示。

```python
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']='SimHei'
plt.rcParams['axes.unicode_minus']=False 
```

```python
# 用图表进行展示
plt.style.use('ggplot')
fig=plt.figure(figsize=(20,8))
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.bar(x='类型',height='基金数量',data=group_number)
for a,b in zip(range(len(group_number.类型)),group_number.基金数量):
    plt.text(a,b,b,ha='center',va='bottom',fontsize=16)
plt.title('各类基金数量',fontdict={'fontsize':20})
```

# 4. 据华夏混合成长基金三年内的单位净值数据进行预测
基于RNN能够很好地处理具有序列特性的数据，挖掘数据中的时序信息，本小组在对基金净值进行建模预测时，所选用的深度学习模型是神经网络中的RNN和LSTM。RNN在进行学习预测时，每一个样本的输出层不仅取决于当前样本的输入层，还取决于前一个样本的相关数据，这使得可以通过RNN预测具有时序特性的数据，即前一个样本值会影响后一个样本值的数据。LSTM是RNN的更高级算法，其具有“门控装置”，相比于普通RNN算法会储存之前所有时刻的数据，LSTM会根据机器学习结果进行有选择的储存，从而筛选出更能够预测其趋势的信息进行预测。

本小组在进行预测时选用了两种不同的方式，并通过误差进行了对比，选择了其中一个更符合的模型作为最终的结果。

## 4.1 方法一：BP神经网络
在这种方式中，本小组将前500个数据作为训练集用以机器学习，将后232条数据作为测试集来查看预测的效果。为了更加方便进行学习和预测，本小组利用特征缩放将训练集和测试集的数据缩放到（-1,1）之间。之后划分输入数据和输出数据，其中输入数据不包含最后一项，输出数据不包含第一项。该部分代码如下：

```python
#划分训练集和测试集
train=list11[:500]
test=list11[500:]
```

```python
scaler = MinMaxScaler(feature_range=(-1, 1))
```

```python
train_sc = scaler.fit_transform(np.array(train).reshape(-1,1))
test_sc = scaler.fit_transform(np.array(test).reshape(-1,1))
```

```python
X_train = train_sc[:-1]
y_train = train_sc[1:]
X_test = test_sc[:-1]
y_test = test_sc[1:]
```

将数据进行完特征缩放后，创建BP神经网络模型，利用Keras中的Sequential模型，通过.add()方法添加层，其输入维度为1层，激活函数设置为整流函数relu，并通过compile方法完成学习过程的配置，损失函数是mean_squared_error即均方误差越小越好，优化器是Adam。同时设置检测到loss停止改进时，结束训练，如果经过数个周期结果依旧没有改进，同样停止。该BP神经网络模型的训练时间为100个周期，每次用1个样本进行训练。具体代码如下：

```python
#定义模型神经网络模型
nn_model = Sequential()
nn_model.add(Dense(12, input_dim=1, activation='relu'))
nn_model.add(Dense(1))
nn_model.compile(loss='mean_squared_error', optimizer='adam')
early_stop = EarlyStopping(monitor='loss', patience=2, verbose=1)
history = nn_model.fit(X_train, y_train, epochs=100, batch_size=1, verbose=1, callbacks=[early_stop], shuffle=False)
```

```python
epochs=range(1,100)
plt.plot(history.history['loss'],'b',label='Train acc',color='black')
plt.title('bpmodel_loss')
plt.legend()
plt.show()
```

该BP神经网络模型通过迭代过程即梯度下降来优化学习过程，并非需要进行100个周期，其在第9个周期即已停止，说明在第9个周期时，其均方误差已足够小。为了进一步展示均方误差，本小组对每个周期误差的变化利用matplotlib进行了可视化（见图bpmodel_loss）。从图四中可以得知，通过不停地梯度下降优化，模型的损失在不断减小，最终低至0.0059，说明模型的拟合程度较好。同时，利用sklearn.metrics中的r2_score可以更好地衡量改模型的拟合优度，训练集的r2_score为0.927，测试集的r2_score为0.941，说明该模型对数据的解释程度均达到了90%以上，能够很好地对华夏混合成长基金三年内的数据进行预测。

```python
y_pred_test_nn = nn_model.predict(X_test)
y_train_pred_nn = nn_model.predict(X_train)
print("The R2 score on the Train set is:\t{:0.3f}".format(r2_score(y_train, y_train_pred_nn)))
print("The R2 score on the Test set is:\t{:0.3f}".format(r2_score(y_test, y_pred_test_nn)))
```

为了更加直观的展示模型对于华夏混合成长基金历史单位净值数据的解释力度，本小组利用matplotlib将测试集中的模型预测值与真实值画在一张图上方便对比研究（见图NN's Prediction）。由图五我们可以直观地看出，在测试集的231条数据中，除去开始50条左右的数据，其后数据的预测值与真实值均相差不大，总体的变化趋势也几乎一致，说明该模型的拟合优度较好，具备一定的对华夏混合成长基金单位净值变化的预测能力。

## 4.2 方法二：循环预测模型（LSTM）
循环预测是一种适用于预测具有周期性特征的数据，为了探究华夏混合成长基金的历史单位净值是否具有一定周期性并进行未来单位净值变化的预测，本小组选取循环预测模型（LSTM）作为第二种机器学习方式。

在这种方式中，本小组利用自定义函数的方法构造数据集。自定义构造了函数create_dataset，将数据集划分为训练集和测试集，每次用前72个数据作为训练集，来预测后12个数据，一共预测288次，如此来达到循环预测的效果。在这部分数据集处理时，依然对数据进行了特征缩放，缩放为（0，1）的数据。最后对数据进行了转换以满足keras。

处理好数据集之后，构建相关机器学习模型，利用.add()来添加层，利用LSTM构建模型，损失函数为mean_squared_error，优化器为adam。

因为利用该模型每次只能预测12个数据，但是最终预测288个数据，故采用循环预测的思路。每次预测的12个数据，添加到数据集中充当预测x，然后在预测新的12个y，再添加到预测x列表中，如此往复预测出288个点。

该部分代码如下：

```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error

#读取数据
data = pd.DataFrame(list11)
dataf = data.values[0:-288]
def create_dataset(dataset, timesteps=36,predict_size=6):#构造数据集
    datax=[]#构造x
    datay=[]#构造y
    for each in range(len(dataset)-timesteps - predict_steps):
        x = dataset[each:each+timesteps,0]
        y = dataset[each+timesteps:each+timesteps+predict_steps,0]
        datax.append(x)
        datay.append(y)
    return datax, datay#np.array(datax),np.array(datay)
#构造train and predict
scaler = MinMaxScaler(feature_range=(0,1))
dataf = scaler.fit_transform(dataf)
train = dataf.copy()
timesteps = 72#构造x，为72个数据,表示每次用前72个数据作为一段
predict_steps = 12#构造y，为12个数据，表示用后12个数据作为一段
length = 288#预测多步，预测288个数据，每次预测12个
trainx, trainy = create_dataset(train, timesteps, predict_steps)
trainx = np.array(trainx)
trainy = np.array(trainy)

#变换
trainx = np.reshape(trainx,(trainx.shape[0],timesteps,1))#变换shape,以满足keras
#lstm training
model = Sequential()
model.add(LSTM(128,input_shape=(timesteps,1),return_sequences= True))
model.add(Dropout(0.5))
model.add(LSTM(128,return_sequences=True))
#model.add(Dropout(0.3))
model.add(LSTM(64,return_sequences=False))
#model.add(Dropout(0.2))
model.add(Dense(predict_steps))
model.compile(loss="mean_squared_error",optimizer="adam")
kk=model.fit(trainx,trainy, epochs= 50, batch_size=200)
#predict
#因为每次只能预测12个数据，但是我要预测288个数据，所以采用的就是循环预测的思路。每次预测的12个数据，添加到数据集中充当预测x，然后在预测新的12个y，再添加到预测x列表中，如此往复。最终预测出288个点。
predict_xlist = []#添加预测x列表
predict_y = []#添加预测y列表
predict_xlist.extend(dataf[dataf.shape[0]-timesteps:dataf.shape[0],0].tolist())#已经存在的最后timesteps个数据添加进列表，预测新值(比如已经有的数据从1,2,3到288。现在要预测后面的数据，所以将216到288的72个数据添加到列表中，预测新的值即288以后的数据）
while len(predict_y) < length:
    predictx = np.array(predict_xlist[-timesteps:])#从最新的predict_xlist取出timesteps个数据，预测新的predict_steps个数据（因为每次预测的y会添加到predict_xlist列表中，为了预测将来的值，所以每次构造的x要取这个列表中最后的timesteps个数据词）
    predictx = np.reshape(predictx,(1,timesteps,1))#变换格式，适应LSTM模型
    #print("predictx"),print(predictx),print(predictx.shape)
    #预测新值
    lstm_predict = model.predict(predictx)
    #predict_list.append(train_predict)#新值y添加进列表，做x
    #滚动预测
    #print("lstm_predict"),print(lstm_predict[0])
    predict_xlist.extend(lstm_predict[0])#将新预测出来的predict_steps个数据，加入predict_xlist列表，用于下次预测
    # invert
    lstm_predict = scaler.inverse_transform(lstm_predict)
    predict_y.extend(lstm_predict[0])#预测的结果y，每次预测的12个数据，添加进去，直到预测288个为止
    #print("xlist", predict_xlist, len(predict_xlist))
    #print(lstm_predict, len(lstm_predict))
    #print(predict_y, len(predict_y))
#error

y_ture = np.array(data.values[-288:])
train_score = np.sqrt(mean_squared_error(y_ture,predict_y))
print("train score RMSE: %.2f"% train_score)
y_predict = pd.DataFrame(predict_y,columns=["predict"])
y_predict.to_csv("y_predict_LSTM.csv",index=False)
#plot
plt.plot(y_ture,c="g")
plt.plot(predict_y, c="r")
plt.show()
```
